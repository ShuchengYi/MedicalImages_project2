```python
import sys
import os
from pathlib import Path

project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.append(project_root)

GNDTRUTH = Path('../data/raw/COMMON_images_masks/')
RAWIMGS = Path('../data/raw/GROUP_images/')
DATA = (RAWIMGS.parent).parent
MASKS = DATA/'masks'
LINREGS = DATA/'linear_registrations'

import SimpleITK as sitk
from utils import *
from linear_registration import register_linear
```

```python
# Import all images and masks and obtain arrays
mov_names = ['g1_54', 'g1_55', 'g1_56']
fix_names = ['common_40', 'common_41', 'common_42']
fix_imgs = [sitk.ReadImage(GNDTRUTH/(name+'_image.nii.gz')) for name in fix_names]
mov_imgs = [sitk.ReadImage(RAWIMGS/(name+'_image.nii.gz')) for name in mov_names]
mov_msks = [sitk.ReadImage(MASKS/(name+'_mask.nii.gz')) for name in mov_names]
reg_HU_imgs = [[sitk.ReadImage(LINREGS/'with_HUmask'/(mov_n+'_linreg_'+fix_n+'_img.nii.gz')) for mov_n in mov_names] for fix_n in fix_names]
reg_HU_msks = [[sitk.ReadImage(LINREGS/'with_HUmask'/(mov_n+'_linreg_'+fix_n+'_msk.nii.gz')) for mov_n in mov_names] for fix_n in fix_names]
reg_sl_imgs = [[sitk.ReadImage(LINREGS/'with_regmask'/(mov_n+'_linreg_'+fix_n+'_img.nii.gz')) for mov_n in mov_names] for fix_n in fix_names]
reg_sl_msks = [[sitk.ReadImage(LINREGS/'with_regmask'/(mov_n+'_linreg_'+fix_n+'_msk.nii.gz')) for mov_n in mov_names] for fix_n in fix_names]

fix_imgs_arr = []
mov_imgs_arr = []
mov_msks_arr = []
reg_HU_imgs_arr = []
reg_HU_msks_arr = []
reg_sl_imgs_arr = []
reg_sl_msks_arr = []
```

```python
for idx, n in enumerate(fix_names):
    fix_imgs_arr.append(sitk.GetArrayFromImage(fix_imgs[idx]))
    mov_imgs_arr.append(sitk.GetArrayFromImage(mov_imgs[idx]))
    mov_msks_arr.append(sitk.GetArrayFromImage(mov_msks[idx]))
```

```python
for idx, n in enumerate(fix_names):
    reg_HU_imgs_arr.append([sitk.GetArrayFromImage(im) for im in reg_HU_imgs[idx]])
    reg_HU_msks_arr.append([sitk.GetArrayFromImage(im) for im in reg_HU_msks[idx]])
    reg_sl_imgs_arr.append([sitk.GetArrayFromImage(im) for im in reg_sl_imgs[idx]])
    reg_sl_msks_arr.append([sitk.GetArrayFromImage(im) for im in reg_sl_msks[idx]])
```
